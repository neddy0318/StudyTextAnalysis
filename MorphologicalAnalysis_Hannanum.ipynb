## Hannanum 형태소 분석

### 1. Hannanum 형태소 분석

&nbsp;&nbsp;&nbsp; -한나눔은 한국어 형태소 분석기로, 자바로 작성된 오픈 소스 형태소 분석기입니다.

&nbsp;&nbsp;&nbsp; -형태소 분석, 명사 추출, 품석 태깅 등의 작업을 수행할 수 있습니다.

&nbsp;&nbsp;&nbsp; -한나눔은 자음과 모음 단위로 분석하는 '음절 분리 방식'을 기반으로 형태소를 추출하고 품사를 태깅합니다.

### 2. haennanum.nouns()

&nbsp;&nbsp;&nbsp; -hannanum.nouns()는 형태소 분석기의 메소드 중 하나로, 한국어 문장에서 명사(nouns)를 추출합니다.

&nbsp;&nbsp;&nbsp; -명사를 추출해서 리스트로 반호나합니다. 이를 통해 문장의 주요 주제나 내용을 파악할 수 있습니다.


### 3. 실습 및 실전에서 써보기

&nbsp;&nbsp;&nbsp; -참고 자료 : https://blog.naver.com/ritardo76/223405415688

```python
import pandas as pd
```

```python
from konlpy.tag import Hannanum

hannanum = Hannanum()
sentence = "나는 오늘 한국어 형태소 분석기인 Hannanum에 대해 공부하고 있습니다. 실전에 적용하려 하니 쉽진 않은 것 같네요. ㅠㅠ"
nouns = hannanum.nouns(sentence)
print(nouns)

#'ㅠㅠ'까지 잘 캐치해냅니다.
```

위와 같이 형태소 분석을 해봤으니, 본격적으로 실전에 적용해 보겠습니다.

```python
#'비건'키워드 뉴스데이터 불러오기

news1 = pd.read_excel('C:/Users/neddy/Documents/GitHub/K-VeganTourism/SSH/비건뉴스_20200722-20220721.xlsx')
news2 = pd.read_excel('C:/Users/neddy/Documents/GitHub/K-VeganTourism/SSH/비건뉴스_20220722-20240721.xlsx')
```

```python
vegan_news = pd.concat([news1, news2], axis=0)
```

```python
vegan_news.head(3)
```

실전에 맞게 코드 응용하기

```python
all_nouns = []

for news in vegan_news['본문']:
    nouns = hannanum.nouns(news)
    all_nouns.extend(nouns)

noun_counts = Counter(all_nouns)

noun_freq_df = pd.DataFrame(noun_counts.items(), columns=['Noun', 'Frequency'])

noun_freq_df = noun_freq_df.sort_values(by='Frequency', ascending=False).reset_index(drop=True)
```
